```python
import numpy as np
import cv2 as cv 
from keras.models import load_model
from keras.preprocessing import image
import  matplotlib.pyplot as plt
import os
```

```python
#image laoder
def load_image (img_path, img_size, show=False):
    img = image.load_img(img_path, target_size=img_size)
    img_tensor = image.img_to_array(img)
    img_tensor = np.expand_dims(img_tensor, axis=0)      #expanding image tensor
    img_tensor = img_tensor /255.           # scaling the image_T

    if show:
        plt.imshow(img_tensor[0])
        plt.axis('off')
        plt.show()
    return img_tensor

img_size = (300, 300)
img_path = 'inference image from medscape.jpg'
model_path = 'tb_model'

```

```python

## load img
model = load_model(model_path)
model.summary()
```

```python
#loading image
img = load_image(img_path, img_size, show=True)
img
```

```python
classes = ['Normal', 'Tuberculosis']
pred = model.predict(img)
```
